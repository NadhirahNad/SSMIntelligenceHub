### LLP

```python
import csv
import re

def clean_text(text):
    text = text.replace('–', '-').replace('—', '-')
    text = text.replace('-LRB-', '(').replace('-RRB-', ')')
    text = re.sub(r'\s?-\s?', '-', text)
    text = text.replace('[', '(').replace(']', ')')
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)
    text = re.sub(r'\n{2,}', '\n', text)
    return text.strip()

def parse_document(text):
    lines = text.splitlines()
    data = []

    current_part = "General"
    current_section = None
    current_subsection = None
    current_content = []

    i = 0
    while i < len(lines):
        line = lines[i].strip()
        if not line:
            i += 1
            continue

        # Detect PART
        part_match = re.match(r'^(New\s+)?Part\s+([IVXLCDM]+[a-zA-Z]?)', line, re.IGNORECASE)
        if part_match:
            current_part = line
            i += 1
            if i < len(lines) and lines[i].strip():
                current_part += ": " + lines[i].strip()
            current_subsection = None
            i += 1
            continue

        # Detect Section line
        section_match = re.match(r'^(Section\s+)?(\d+[a-zA-Z]*)\.\s*(.*)', line)
        if section_match:
            # Save the previous section
            if current_section and current_content:
                data.append({
                    'Part': current_part,
                    'Subsection': current_subsection,
                    'Section': current_section,
                    'Content': ' '.join(current_content).strip()
                })

            section_number = section_match.group(2)
            section_title = section_match.group(3).strip()
            current_section = f"Section {section_number}"
            current_content = [section_title] if section_title else []

            # Look back 1 line for possible subsection title
            if i > 0:
                prev_line = lines[i - 1].strip()
                if (
                    prev_line and
                    not re.match(r'^(Section\s+)?\d+[a-zA-Z]*\.', prev_line) and
                    not re.match(r'^(New\s+)?Part\s+([IVXLCDM]+[a-zA-Z]?)', prev_line, re.IGNORECASE)
                ):
                    current_subsection = prev_line
                else:
                    current_subsection = None
            else:
                current_subsection = None

            # Capture multi-line content
            j = i + 1
            while j < len(lines):
                next_line = lines[j].strip()
                next_section_match = re.match(r'^(Section\s+)?(\d+[a-zA-Z]*)\.\s*', next_line)
                next_part_match = re.match(r'^(New\s+)?Part\s+([IVXLCDM]+[a-zA-Z]?)', next_line, re.IGNORECASE)
                if next_line and not next_section_match and not next_part_match:
                    current_content.append(next_line)
                    j += 1
                else:
                    break
            i = j
            continue

        # Regular content line
        current_content.append(line)
        i += 1

    # Final flush
    if current_section and current_content:
        data.append({
            'Part': current_part,
            'Subsection': current_subsection,
            'Section': current_section,
            'Content': ' '.join(current_content).strip()
        })

    return data

def write_to_csv(data, output_file):
    with open(output_file, 'w', newline='', encoding='utf-8') as f:
        writer = csv.DictWriter(f, fieldnames=['Part', 'Subsection', 'Section', 'Content'])
        writer.writeheader()
        writer.writerows(data)

def process_file(input_txt_path, output_csv_path):
    with open(input_txt_path, 'r', encoding='utf-8') as f:
        text = f.read()

    cleaned = clean_text(text)
    structured = parse_document(cleaned)
    write_to_csv(structured, output_csv_path)
    print(f"✅ CSV created: {output_csv_path}")

# File paths
input_txt_path = r'C:\\Users\\Nadhirah\\OneDrive\\Documents\\KAKAK\\NAD SSM\\BSD3724 DSP2\\OUTPUT\\Act LLP 2024.txt'
output_csv_path = r'C:\\Users\\Nadhirah\\OneDrive\\Documents\\KAKAK\\NAD SSM\\BSD3724 DSP2\\OUTPUT\\Chunk LLP 2024.csv'

# Run
process_file(input_txt_path, output_csv_path)


```

## POS Tagging

```python
import pandas as pd
import spacy

nlp = spacy.load("en_core_web_sm")

def pos_tag_text(text):
    doc = nlp(text)
    return [(token.text, token.pos_) for token in doc]

def process_pos(act_name, input_csv_path, output_csv_path):
    df = pd.read_csv(input_csv_path)

    pos_tags = []

    print(f"\nPerforming POS tagging for {act_name}...")
    for content in df['Content']:
        pos = pos_tag_text(str(content))
        pos_tags.append(pos)

    df['POS_Tags'] = pos_tags
    df.to_csv(output_csv_path, index=False)
    print(f"POS tagging for {act_name} complete. File saved to: {output_csv_path}")

if __name__ == "__main__":
    # Details for both Acts
    documents = [
        {
            "act_name": "ROB Act 1956",
            "input": r"C:\\Users\\Nadhirah\\OneDrive\\Documents\\KAKAK\\NAD SSM\\BSD3724 DSP2\\OUTPUT\\Chunk ROB 1956 EDITED.csv",
            "output": r"C:\\Users\\Nadhirah\\OneDrive\\Documents\\KAKAK\\NAD SSM\\BSD3724 DSP2\\OUTPUT\\pos_rob.csv"
        },
        {
            "act_name": "ROC Act 2016",
            "input": r"C:\\Users\\Nadhirah\\OneDrive\\Documents\\KAKAK\\NAD SSM\\BSD3724 DSP2\\OUTPUT\\Chunk ROC 2016 EDITED.csv",
            "output": r"C:\\Users\\Nadhirah\\OneDrive\\Documents\\KAKAK\\NAD SSM\\BSD3724 DSP2\\OUTPUT\\pos_roc.csv"
        }
    ]

    for doc in documents:
        process_pos(doc["act_name"], doc["input"], doc["output"])



```

## NER Tagging

```python
import pandas as pd
from transformers import pipeline

# Load the NER pipeline using LegalBERT
ner_pipeline = pipeline("ner", model="nlpaueb/legal-bert-small-uncased", aggregation_strategy="simple")

def ner_text(text):
    ner_results = ner_pipeline(text)
    return [(ent['word'], ent['entity_group']) for ent in ner_results]

def add_ner_and_save_new(act_name, input_csv_path, output_csv_path):
    df = pd.read_csv(input_csv_path)

    ner_entities = []

    print(f"Performing Named Entity Recognition for {act_name}...")
    for content in df['Content']:
        ner = ner_text(str(content))
        ner_entities.append(ner)

    df['NER_Entities'] = ner_entities
    df.to_csv(output_csv_path, index=False)
    print(f"NER for {act_name} complete. File saved to: {output_csv_path}")

if __name__ == "__main__":
    documents = [
        {
            "act_name": "ROB Act 1956",
            "input": r"C:\\Users\\Nadhirah\\OneDrive\\Documents\\KAKAK\\NAD SSM\\BSD3724 DSP2\\OUTPUT\\pos_rob.csv",
            "output": r"C:\\Users\\Nadhirah\\OneDrive\\Documents\\KAKAK\\NAD SSM\\BSD3724 DSP2\\OUTPUT\\posner_rob.csv"
        },
        {
            "act_name": "ROC Act 2016",
            "input": r"C:\\Users\\Nadhirah\\OneDrive\\Documents\\KAKAK\\NAD SSM\\BSD3724 DSP2\\OUTPUT\\pos_roc.csv",
            "output": r"C:\\Users\\Nadhirah\\OneDrive\\Documents\\KAKAK\\NAD SSM\\BSD3724 DSP2\\OUTPUT\\posner_roc.csv"
        },
#        {
#            "act_name": "LLP Act 2024",
#            "input": r"C:\\Users\\Nadhirah\\OneDrive\\Documents\\KAKAK\\NAD SSM\\BSD3724 DSP2\\OUTPUT\\pos_llp.csv",
#            "output": r"C:\\Users\\Nadhirah\\OneDrive\\Documents\\KAKAK\\NAD SSM\\BSD3724 DSP2\\OUTPUT\\posner_llp.csv"
#        }
    ]

    for doc in documents:
        add_ner_and_save_new(doc["act_name"], doc["input"], doc["output"])


```

```python
import pandas as pd
from neo4j import GraphDatabase

# === Neo4j Connection ===
uri = "neo4j://127.0.0.1:7687"
username = "neo4j"
password = "ssm_database"
driver = GraphDatabase.driver(uri, auth=(username, password))

# === Load CSV ===
file_path = r"C:\\Users\\Nadhirah\\OneDrive\\Documents\\KAKAK\\NAD SSM\\BSD3724 DSP2\\OUTPUT\\Chunk ROB 1956 EDITED.csv"
df = pd.read_csv(file_path)

# === Clean header ===
df.columns = [col.strip() for col in df.columns]

# === Insert Function ===
def insert_rob(tx, rows):
    for _, row in rows.iterrows():
        part = str(row.get('Part', '')).strip()
        section = str(row.get('Section', '')).strip()
        content = str(row.get('Content', '')).strip()

        if not section:
            continue

        params = {
            "part": part,
            "section": section,
            "content": content
        }

        query = """
        MERGE (a:Act {name: "ROB 1956"})
        """

        if part:
            query += """
            MERGE (a)-[:HAS_PART]->(p:Part {name: $part, act: "ROB 1956"})
            MERGE (p)-[:HAS_SECTION]->(s:Section {name: $section, part: $part, act: "ROB 1956"})
            """
        else:
            query += """
            MERGE (a)-[:HAS_SECTION]->(s:Section {name: $section, act: "ROB 1956"})
            """

        query += """
        MERGE (s)-[:HAS_CONTENT]->(c:Content {text: $content})
        """

        tx.run(query, params)

# === Batch Insert ===
batch_size = 100
with driver.session() as session:
    for i in range(0, len(df), batch_size):
        batch = df.iloc[i:i + batch_size]
        session.write_transaction(insert_rob, batch)
        print(f"Inserted batch {i} to {i + len(batch) - 1}")

driver.close()
print("✅ All data inserted.")
```
